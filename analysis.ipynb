```python
import pandas as pd 
import yfinance as yf
import datetime as dt
```

First, the data collection <br><br>
Get the top 50 companies in SP500 index

```python
# list of tickers of top 100 companies in SP500 by market cap
tickers = [
    "MSFT", "AAPL", "NVDA", "GOOG", "GOOGL", "AMZN", "META", "BRK-B", "LLY", "AVGO", 
    "TSLA", "JPM", "V", "WMT", "XOM", "MA", "PG", "UNH", "HD", "MRK", "CVX", 
    "ABBV", "KO", "PEP", "ORCL", "PFE", "COST", "TMO", "ADBE", "MCD", "CRM", 
    "NFLX", "ACN", "TXN", "NEE", "AMD", "ABT", "PM", "LIN", "UPS", "INTU", 
    "VZ", "QCOM", "HON", "MS", "GS", "AMT", "SBUX", "CAT", "LOW", "DE", 
    "MDLZ", "AMGN", "LMT", "BKNG", "DHR", "RTX", "SPGI", "BLK", "SCHW", "CI", 
    "SYK", "T", "ELV", "PLD", "TJX", "MMM", "MO", "GILD", "ADP", "ISRG", 
    "PNC", "ITW", "USB", "MRNA", "AXP", "FISV", "MU", "CB", "MDT", "MMC", 
    "SO", "ADI", "PGR", "C", "NKE", "ETN", "VRTX", "COP", "PANW", "CL", 
    "CHTR", "TGT", "DUK", "EQIX", "KMB", "ZTS", "SHW", "EOG", "REGN", "APD"
]
```

```python
num_tickers = 10

# get todays date
date_today = dt.datetime.today().date()

# Download historical data for all tickers since January 1, 2000
data = yf.download(tickers[:num_tickers], start="2005-01-01", end=date_today, group_by="ticker")

data
```

```python
# leave only the closing price of each ticker

close_prices = pd.DataFrame()

for ticker in tickers[:10]:
    close_prices[ticker] = data[ticker]["Close"]

# close_prices=close_prices.reset_index()
close_prices.dropna(inplace=True)
```

After we have the data let's analyze it

```python
close_prices
```

```python
import plotly.io as pio
# Set the default rendering engine to static image
pio.renderers.default = "browser"
```

```python
import plotly.graph_objs as go
import plotly.express as px

fig = go.Figure()

for company in close_prices.columns[1:]:
    fig.add_trace(go.Scatter(x=close_prices.index, y=close_prices[company],
                             mode='lines',
                             name=company,
                             opacity=0.5))

fig.update_layout(
    title='Stock Price Trends of SP500 Companies',
    xaxis_title='Date',
    yaxis_title='Closing Price (INR)',
    xaxis=dict(tickangle=45), 
    legend=dict(
        x=1.05,
        y=1,
        traceorder="normal",
        font=dict(size=10),
        orientation="v"
    ),
    margin=dict(l=0, r=0, t=30, b=0),  
    hovermode='x',
    template='plotly_white'
)

fig.show()
```

Calculate the percentage change for each company between each day

```python
all_companies = close_prices.columns

volatility_all_companies = close_prices[all_companies].std()

volatility_all_companies.sort_values(ascending=False).head(10)
```
